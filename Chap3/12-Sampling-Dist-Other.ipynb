We have encountered situations where constructing sampling distributions
for means and proportions are helpful. In addition to being useful
summaries of data, both means and proportions are linear combinations of
independent random variables.

-   If $X_1, X_2, \ldots , X_n$ are independent and identically
    distributed (i.i.d.) random variables (independently selected from
    the same population), then the sample mean is a linear combination
    of random variables,

$$\bar{X} = \frac{X_1 + X_2 + \ldots + X_n}{n}.$$

-   If $X \sim \mbox{Binom}(n, p)$, then the sample proportion is a
    multiple of one random variable,

$$\widehat{P} = \frac{X}{n}.$$ - Thus for proportions and means, we can
derive the Central Limit Theorem (CLT) to describe the mean and standard
error (spread) of sampling distributions applying properties of expected
value and variance of linear combinations of independent random
variables $X$ and $Y$, namely:

$$E(aX + bY) = aE(X) + bE(Y) \qquad \mbox{and} \qquad \mbox{Var}(aX + bY) = a^2\mbox{Var}E(X) + b^2 \mbox{Var}E(Y).$$
There many other types and combinations of statistics that are
interesting to study to other situations. We can consider more advanced
questions. **Even if we cannot derive formulas that describe the mean
and variance of a sampling distribution for certain statistics, we can
always construct a sampling distribution using a simulation.**

# Simulation for Sampling Distributions

------------------------------------------------------------------------

We used the same general process for constructing sampling distributions
for [means](10-Sampling-Dist-Mean.qmd) as with for
[proportions](11-Sampling-Dist-Prop.qmd). We can apply the same process
to create a sampling distribution for any statistic or combination of
statistics.

1.  Pick a random sample size $n$ from population $X$.
2.  Calculate meaning statistic(s) for the sample.
3.  Repeat and generate many random samples (each size $n$).

For example, we considered the [sampling distribution for the mean wait
time between trains](10-Sampling-Dist-Mean.qmd) when the wait time
between trains was exponentially distributed with a mean of $4$ minutes
between trains. The rate parameter $\lambda = \frac{1}{4}$. We sample
from the population $Y \sim \mbox{Exp} \left( \frac{1}{4} \right)$ which
is skewed to the right.

```python
# creates an empty vector to store results
samp.mean <- numeric(10000)
n <- 50

# a for loop that generates 10000 random samples 
# each size n=50, and calculates the sample mean.
for (i in 1:10000)
{
  time.sample <- rexp(50, 1/4)  # Step 1: Randomly picks 50 values from Exp(1/4)
  samp.mean[i] <- mean(time.sample)  # Step 2: calculate sample mean
}
```

## Graphical Summary of Results of Simulation

------------------------------------------------------------------------

```python
par(mfrow=c(1,2))  # arrange plots side by side

# plot population
y <- seq(0, 6, 0.01) 
plot(y, dexp(y), type = "l",
     xlab = "Y, wait time (in min)",
     main = "Population, Y ~ Exp(1/4)")

# plot the sampling distribution of means
hist(samp.mean, 
     breaks = 20,
     xlab = "Mean Wait Time (n=50)",
     main = "Dist. of Sample Means")
```

# Exploring the Theory

------------------------------------------------------------------------

In the previous examples involving sample median and variance, we are
able to analyze properties of the sampling distribution of those
statistics by simulating the creation of a sampling distribution. We can
always use simulations to approximate sampling distributions, even if we
do not have a formal theorem and/or formulas for the mean and standard
error of the sampling distribution.

With some statistics, such as sample means and sample proportions, we
can use properties of random variables to derive theory and formulas
that describe properties of the sampling distribution, and these cases
technology is not required to simulate the sampling distribution. We
explore some additional statistics, namely minimum and maximum values of
a sample, and derive formulas to describe their sampling distributions.

## Sampling Distribution for the Maximum

------------------------------------------------------------------------

Frequently we are interested in the minimum value in a random sample,
denoted $X_{\rm{min}}$, or maximum value, denoted $X_{\rm{max}}$ of
random sample. For example, if we pick a random sample of five values,
$\left\{ 11, 18, 2, 31, 25 \right}$, then we have the statistics
$X_{\rm{min}} = 2$ and $X_{\rm{max}} = 31$.

A <span style="color:dodgerblue">**sampling distribution for the sample
maximum**</span> would be the distribution of maximum values obtained
from samples when we independently pick many random samples, each size
$n$, from the same population.

F_{X_{\rm{max}}} (a) &= P(X_1 \leq a, X_2 \leq a, \ldots, X_n \leq a) & \mbox{Explanation 1} \\

&= F(a) \cdot F(a) \cdot \ldots \cdot F(a) & \mbox{Explanation 4}\\
&= \big( F(a) \big)^n
\end{aligned}$$

### Summarizing Results for $X_{\rm{max}}$

------------------------------------------------------------------------

Let $F$ denote the cumulative distribution function (cdf) for a
continuous random variable $X$, and let $X_{\rm{max}$ denote the maximum
of a random sample $X_1,X_2, \ldots, X_n$, where each $X_i$ is
independently picked from population $X$. Then $X_{\rm{max}$ will have:

-   Cumulative distribution function
    $F_{X_{\rm{max}}} (a) = \big() F(a) \big)^n$.
-   Probability density function
    $f_{X_{\rm{max}}} = n \left( F(a) \right)^{n-1} f(a$)

## Simulating a Distribution of Sample $\chi^2$

------------------------------------------------------------------------

We use a similar process as with sampling distributions for
[means](10-Sampling-Dist-Mean.qmd) and
[proportions](11-Sampling-Dist-Prop.qmd).

1.  Pick a random sample size $n$ from the population (NYC in 2022).

```python
temp<- sample(c("Black", "Latinx", "White", "Asian/Pacific Islander", "Mixed/Other"),
                      size = 15102,  
                      replace = TRUE, 
                      prob = c(0.23, 0.29, 0.40, 0.14, 0.06))
```

1.  Calculate the chi-square test statistic of the sample as described
    in the previous section.

```python
o1 <- sum(temp == "Black")
o2 <- sum(temp == "Latinx")
o3 <- sum(temp == "White")
o4 <- sum(temp == "Asian/Pacific Islander")
o5 <- sum(temp == "Mixed/Other")

temp.obs <- c(o1, o2, o3, o4, o5)
rel.diff <- (temp.obs - expect)^2 / expect
sample.chi <-sum(rel.diff)
```

1.  Repeat many (1,000) times.

```python
dist.chi <- numeric(1000) # creates an empty vector to store results
n <- 15102  # total number of observed
prop <- c(0.23, 0.29, 0.40, 0.14, 0.06)  # nyc racial demographics
expect <- n * prop  # sample size * population prop

# a for loop that generates 1000 random samples 
for (i in 1:1000)
{
  temp<- sample(c("Black", "Latinx", "White", "Asian/Pacific Islander", "Mixed/Other"),
                      size = n,  
                      replace = TRUE, 
                      prob = c(0.23, 0.29, 0.40, 0.14, 0.06))
  o1 <- sum(temp == "Black")
  o2 <- sum(temp == "Latinx")
  o3 <- sum(temp == "White")
  o4 <- sum(temp == "Asian/Pacific Islander")
  o5 <- sum(temp == "Mixed/Other")
  temp.obs <- c(o1, o2, o3, o4, o5)
  rel.diff <- (temp.obs - expect)^2 / expect
  dist.chi[i] <-sum(rel.diff)
}

hist(dist.chi,  # plot stored sample statistics
     breaks = 20,  # use 20 breaks
     xlab = "Sample Goodness-of-Fit",  # x-axis label
     main = "Sampling Distribution of Chi-Squared Statistics",  # main label
     freq = FALSE)  # plot relative frequencies (density) on y-axis
```
